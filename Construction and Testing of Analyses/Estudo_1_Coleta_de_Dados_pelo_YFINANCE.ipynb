```python
# 📌 Importar bibliotecas
import pandas as pd
import yfinance as yf
from google.colab import files
```

```python
# 📌 Passo 1: Carregar o arquivo CSV da Interactive Brokers
print("📂 Faça o upload do arquivo CSV da IBKR...")
uploaded = files.upload()  # Espera o usuário carregar o arquivo
```

```python
# Identificar o nome do arquivo carregado
file_name = list(uploaded.keys())[0]
```

```python
# Carregar o arquivo CSV
df_ibkr = pd.read_csv(file_name, delimiter=';')
```

```python
# 📌 Exibir as primeiras linhas do arquivo para conferir a estrutura
print("📊 Visualizando os primeiros dados do arquivo carregado:")
df_ibkr.head()
```

```python
# Criar lista para armazenar os dados coletados
dados_ativos_yf = []
```

```python
# 📌 Passo 2: Iterar sobre os ativos da Interactive Brokers e buscar dados no Yahoo Finance
for index, row in df_ibkr.iterrows():
    symbol = row["symbol"].strip()

    try:
        # 📌 Buscar dados no Yahoo Finance
        ativo = yf.Ticker(symbol)
        info = ativo.info

        # 📌 Criar um dicionário para armazenar os dados do ativo, mantendo as colunas do df_ibkr
        ativo_data = {
            "symbol": symbol,
            "description": row.get("description", "N/A"),
            "currency": row.get("currency", "N/A"),
            "product": row.get("product", "N/A"),
            "region": row.get("region", "N/A"),
            "Nome": info.get("longName", "N/A"),
            "Preço Atual": info.get("currentPrice", "N/A"),
            "P/E Ratio": info.get("trailingPE", "N/A"),
            "Dividend Yield": info.get("dividendYield", "N/A"),
            "Beta": info.get("beta", "N/A"),
            "Volume": info.get("volume", "N/A")
        }

        # Adicionar os dados à lista
        dados_ativos_yf.append(ativo_data)

    except Exception as e:
        print(f"❌ Erro ao buscar {symbol}: {e}")

# 📌 Criar DataFrame com os dados coletados, mantendo as colunas do df_ibkr
df_dados_ativos_yf = pd.DataFrame(dados_ativos_yf)

# 📌 Exibir os dados coletados no Google Colab
df_dados_ativos_yf.head(20)
```

```python
# Salvar os dados coletados em um CSV
df_dados_ativos_yf.to_csv("dados_ativos_yfinance.csv", index=False)

# Baixar o arquivo CSV
files.download("dados_ativos_yfinance.csv")
```
