```python
# 📌 Definir o nome do arquivo
file_name = "dados_ativos_yfinance.csv"
```

```python
# 📌 Carregar o CSV no DataFrame
df_ativos = pd.read_csv(file_name)
```

```python
# 📌 Exibir as primeiras linhas para validação
df_ativos.head()
```

```python
# 📌 Remover ativos que não possuem preço atual
df_ativos = df_ativos.dropna(subset=["Preço Atual"])

# 📌 Exibir os dados filtrados
df_ativos.head(20)  # Mostra os primeiros 20 ativos para verificação
```

```python
import yfinance as yf
import pandas as pd
import time
```

```python
# Criar lista para armazenar setores
setores = []
```

```python
# 📌 Definir tempo limite para cada requisição
timeout_seconds = 5  # Tempo máximo de espera por ativo
```

```python
# 📌 Criar um conjunto para evitar buscas repetidas
ativos_processados = set()
```

```python
# 📌 Iterar sobre os ativos
for symbol in df_ativos["symbol"]:
    # Evitar segunda tentativa de acesso ao mesmo ativo
    if symbol in ativos_processados:
        continue

    try:
        # 📌 Limitar tempo de espera para evitar travamentos
        start_time = time.time()

        ativo = yf.Ticker(symbol)  # Buscar ativo
        info = ativo.info  # Obter informações

        # Se o tempo de espera for maior que o limite, pular ativo
        if time.time() - start_time > timeout_seconds:
            print(f"⏳ Timeout ao buscar {symbol}. Pulando ativo...")
            setores.append({"symbol": symbol, "Setor": "Timeout"})
            ativos_processados.add(symbol)
            continue

        # 📌 Obter setor do ativo
        setor = info.get("sector", "N/A")

        # 📌 Armazenar no DataFrame
        setores.append({"symbol": symbol, "Setor": setor})
        ativos_processados.add(symbol)

    except Exception as e:
        print(f"⚠️ Erro ao buscar setor para {symbol}: {e}")
        setores.append({"symbol": symbol, "Setor": "Não encontrado"})
        ativos_processados.add(symbol)
```

```python
# 📌 Criar DataFrame com setores
df_setores = pd.DataFrame(setores)
```

```python
# 📌 Juntar os setores ao DataFrame original
df_ativos = df_ativos.merge(df_setores, on="symbol", how="left")
```

# **Cálculo 1: Volatilidade e Retorno Esperado**
A volatilidade mede o risco de um ativo, enquanto o retorno esperado é a média dos retornos diários.

```python
# Criar um novo DataFrame df_calculos a partir de df_ativos
df_calculos = df_ativos.copy()

df_calculos.head(5)
```

```python
import numpy as np

# 📌 Criar colunas para os cálculos no df_calculos

# 📌 Cálculo do Preço Justo usando P/E Ratio (evitando divisões por zero)
df_calculos["Preço Justo"] = df_calculos["Preço Atual"] / df_calculos["P/E Ratio"].replace(0, np.nan)

# 📌 Cálculo do Dividend Yield Percentual (se estiver em decimal, converter para %)
df_calculos["Dividend Yield (%)"] = df_calculos["Dividend Yield"] * 100

df_calculos.head(50)
```
