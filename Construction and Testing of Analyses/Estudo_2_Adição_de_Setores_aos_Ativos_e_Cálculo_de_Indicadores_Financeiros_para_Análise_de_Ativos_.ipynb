```python
# ðŸ“Œ Definir o nome do arquivo
file_name = "dados_ativos_yfinance.csv"
```

```python
# ðŸ“Œ Carregar o CSV no DataFrame
df_ativos = pd.read_csv(file_name)
```

```python
# ðŸ“Œ Exibir as primeiras linhas para validaÃ§Ã£o
df_ativos.head()
```

```python
# ðŸ“Œ Remover ativos que nÃ£o possuem preÃ§o atual
df_ativos = df_ativos.dropna(subset=["PreÃ§o Atual"])

# ðŸ“Œ Exibir os dados filtrados
df_ativos.head(20)  # Mostra os primeiros 20 ativos para verificaÃ§Ã£o
```

```python
import yfinance as yf
import pandas as pd
import time
```

```python
# Criar lista para armazenar setores
setores = []
```

```python
# ðŸ“Œ Definir tempo limite para cada requisiÃ§Ã£o
timeout_seconds = 5  # Tempo mÃ¡ximo de espera por ativo
```

```python
# ðŸ“Œ Criar um conjunto para evitar buscas repetidas
ativos_processados = set()
```

```python
# ðŸ“Œ Iterar sobre os ativos
for symbol in df_ativos["symbol"]:
    # Evitar segunda tentativa de acesso ao mesmo ativo
    if symbol in ativos_processados:
        continue

    try:
        # ðŸ“Œ Limitar tempo de espera para evitar travamentos
        start_time = time.time()

        ativo = yf.Ticker(symbol)  # Buscar ativo
        info = ativo.info  # Obter informaÃ§Ãµes

        # Se o tempo de espera for maior que o limite, pular ativo
        if time.time() - start_time > timeout_seconds:
            print(f"â³ Timeout ao buscar {symbol}. Pulando ativo...")
            setores.append({"symbol": symbol, "Setor": "Timeout"})
            ativos_processados.add(symbol)
            continue

        # ðŸ“Œ Obter setor do ativo
        setor = info.get("sector", "N/A")

        # ðŸ“Œ Armazenar no DataFrame
        setores.append({"symbol": symbol, "Setor": setor})
        ativos_processados.add(symbol)

    except Exception as e:
        print(f"âš ï¸ Erro ao buscar setor para {symbol}: {e}")
        setores.append({"symbol": symbol, "Setor": "NÃ£o encontrado"})
        ativos_processados.add(symbol)
```

```python
# ðŸ“Œ Criar DataFrame com setores
df_setores = pd.DataFrame(setores)
```

```python
# ðŸ“Œ Juntar os setores ao DataFrame original
df_ativos = df_ativos.merge(df_setores, on="symbol", how="left")
```

# **CÃ¡lculo 1: Volatilidade e Retorno Esperado**
A volatilidade mede o risco de um ativo, enquanto o retorno esperado Ã© a mÃ©dia dos retornos diÃ¡rios.

```python
# Criar um novo DataFrame df_calculos a partir de df_ativos
df_calculos = df_ativos.copy()

df_calculos.head(5)
```

```python
import numpy as np

# ðŸ“Œ Criar colunas para os cÃ¡lculos no df_calculos

# ðŸ“Œ CÃ¡lculo do PreÃ§o Justo usando P/E Ratio (evitando divisÃµes por zero)
df_calculos["PreÃ§o Justo"] = df_calculos["PreÃ§o Atual"] / df_calculos["P/E Ratio"].replace(0, np.nan)

# ðŸ“Œ CÃ¡lculo do Dividend Yield Percentual (se estiver em decimal, converter para %)
df_calculos["Dividend Yield (%)"] = df_calculos["Dividend Yield"] * 100

df_calculos.head(50)
```
